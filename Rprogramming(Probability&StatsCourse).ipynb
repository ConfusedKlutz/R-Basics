```r
# Function to calculate factorial manually
factorial_manual <- function(num) {
  # Check if the number is negative
  if(num < 0) {
    return("Error: Factorial of a negative number doesn't exist.")
  } else if(num == 0 || num == 1) {
    return(1)
  } else {
    # Initialize the result to 1
    result <- 1

    # Loop to multiply numbers from 2 to num
    for(i in 2:num) {
      result <- result * i
    }
    return(result)
  }
}

# Taking input from the user
num <- as.integer(readline(prompt="Enter a number: "))

# Calculating factorial manually
factorial_value <- factorial_manual(num)

# Printing the result
cat("Factorial of", num, "is", factorial_value, "\n")
```

```r
# Taking input from the user
n <- as.integer(readline(prompt="Enter the number of terms: "))

# Initializing the first two terms
fibonacci <- numeric(n)  #vector
fibonacci[1] <- 0
if(n > 1) {
  fibonacci[2] <- 1
}

# Generating the Fibonacci sequence
if(n > 2) {
  for(i in 3:n) {
    fibonacci[i] <- fibonacci[i-1] + fibonacci[i-2]
  }
}

# Printing the sequence
cat("The first", n, "terms of the Fibonacci sequence are:\n")
print(fibonacci)
```

```r
num1 <- as.numeric(readline(prompt="Enter first number: "))
num2 <- as.numeric(readline(prompt="Enter second number: "))
operator <- readline(prompt="Enter operator (+, -, *, /): ")

result <- switch(operator,
"+" = num1+num2,
"-" = num1-num2,
"*" = num1*num2,
"/" = ifelse(num2!=0, num1/num2, "Error: disvision by zero")
)
cat ("result: ", result, "\n")

```
